```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

#Import dataset
dataset = pd.read_csv('file path', sep=";", decimal=',', encoding='latin-1')
dataset.head

#Choose Bundesland and extract rows which list information relevant to it ('Burgenland', 'Kärnten', 
#'Niederösterreich', 'Oberösterreich','Salzburg', 'Steiermark', 'Tirol', 'Vorarlberg', 'Wien')
bundesland = ('Wien')
final_df = dataset.where(dataset['Bundesland'] == bundesland).dropna(how='any')

#Plot Reff with confidence interval
plt.figure(figsize=(15,8))
sns.lineplot(x='Datum', y='R_eff', data=final_df)
plt.fill_between(final_df.Datum, final_df.R_eff_lwr, final_df.R_eff_upr, color='b', alpha=0.2)
plt.xticks(rotation="90")
plt.xticks(final_df['Datum'][::10])  
plt.ylabel('Reff')
plt.ylim([0, 3])
plt.grid()
```
